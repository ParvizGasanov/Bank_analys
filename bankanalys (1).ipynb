```python
df["MINIMUM_PAYMENTS"] = df["MINIMUM_PAYMENTS"].fillna(df["MINIMUM_PAYMENTS"].median())
df["CREDIT_LIMIT"] = df["CREDIT_LIMIT"].fillna(df["CREDIT_LIMIT"].median())
```

```python
features = [
    "BALANCE",
    "PURCHASES",
    "ONEOFF_PURCHASES",
    "INSTALLMENTS_PURCHASES",
    "CASH_ADVANCE",
    "PURCHASES_FREQUENCY",
    "ONEOFF_PURCHASES_FREQUENCY",
    "PURCHASES_INSTALLMENTS_FREQUENCY",
    "CASH_ADVANCE_FREQUENCY",
    "CASH_ADVANCE_TRX",
    "PURCHASES_TRX",
    "CREDIT_LIMIT",
    "PAYMENTS",
    "MINIMUM_PAYMENTS",
    "PRC_FULL_PAYMENT",
    "TENURE"
]
```

```python
X = df[features]
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
inertia = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
```

```python
plt.plot(range(2, 11), inertia, marker="o")
plt.xlabel("Antal kluster")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_scores = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    silhouette_scores.append(score)
```

```python
plt.plot(range(2, 11), silhouette_scores, marker="o")
plt.xlabel("Antal kluster")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Method")
plt.show()
```

```python
for k, score in zip(range(2, 11), silhouette_scores):
    print("Antal kluster:", k, "Silhouette score:", score)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df["Cluster"] = kmeans.fit_predict(X_scaled)
```

```python
df["Cluster"].value_counts()
```

```python
cluster_summary = df.groupby("Cluster")[features].mean().round(2)
cluster_summary
```

```python
segment_names = {
    0: "Kontantuttagskunder",
    1: "Lågaktiva vardagskunder",
    2: "Högspenderande aktiva kunder"
}

df["Segment"] = df["Cluster"].map(segment_names)
```

```python
df[["Cluster", "Segment"]].head(10)
```

```python
df.to_csv("bank_data_with_segments.csv", index=False)
```

## Val av antal kluster

Vi testade flera olika antal kluster med hjälp av Elbow Method och Silhouette Score.
Det högsta silhouette-värdet fick vi vid 3 kluster.
Därför valde vi 3 segment eftersom det både gav ett bra modellresultat och var lätt att tolka ur ett affärsperspektiv.

## Tolkning av segmenten

### Kontantuttagskunder
Detta segment kännetecknas av hög cash advance och relativt låg köpfrekvens. 
Det tyder på kunder som använder kreditkortet mer för kontantuttag än för vanliga köp.

### Lågaktiva vardagskunder
Detta segment består av kunder med lägre aktivitet, lägre köpbelopp och lägre användning av kortet. 
Det är samtidigt den största kundgruppen och kan därför vara viktig för aktiveringskampanjer.

### Högspenderande aktiva kunder
Detta segment har högst köpbelopp och högst aktivitet. 
Det tyder på kunder som använder kortet ofta och är särskilt viktiga för premium- och lojalitetserbjudanden.

## Rekommendationer till marknadschefen

1. Prioritera högspenderande aktiva kunder med premiumerbjudanden, bonusprogram och personliga kampanjer eftersom de har högst köpbelopp och hög aktivitet.

2. Arbeta med aktiveringskampanjer för lågaktiva vardagskunder eftersom de är den största kundgruppen men använder kortet mindre.

3. Följ kontantuttagskunder ur ett risk- och kreditperspektiv eftersom de har hög cash advance och ett annat användningsmönster än övriga segment.
